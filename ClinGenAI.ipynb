```python
import requests, json, time, os

def call_llm(model, prompt):
    url = "https://openrouter.ai/api/v1/chat/completions"
    headers = {
        "Authorization": f"Bearer {os.getenv('OPENROUTER_API_KEY')}",
        "Content-Type": "application/json"
    }
    data = {
        "model": model,
        "messages": [{"role": "user", "content": prompt}],
        "temperature": 0.4
    }

    r = requests.post(url, headers=headers, json=data)

    try:
        response = r.json()
    except:
        print("âŒ Bad JSON:", r.text)
        return None

    # Handle API errors
    if "error" in response:
        print(f"âŒ API Error from {model}: {response['error']}")
        return None

    # Handle malformed results
    try:
        return response["choices"][0]["message"]["content"]
    except KeyError:
        print(f"âŒ Unexpected response from {model}: {response}")
        return None
```

```python
expert_models = {
    "fever_expert":      "deepseek/deepseek-chat",
    "symptom_expert":    "qwen/qwen-2.5-7b-instruct",
    "disease_expert":    "meta-llama/llama-3-8b-instruct",
    "diagnosis_expert":  "mistralai/mistral-7b-instruct",
    "drug_expert":       "mistralai/mistral-nemo",
    "lab_expert":        "openai/gpt-4o-mini",
    "risk_expert":       "anthropic/claude-3.5-haiku"
}
```

```python
expert_prompts = {
    "fever_expert": (
        "Evaluate fever-related clinical reasoning using global guidelines. "
        "Consider threshold values, duration, and epidemiological context. "
        "Reference evidence from:",
        ["WHO", "CDC"]
    ),
    "symptom_expert": (
        "Analyze presenting symptoms and map patterns to possible etiologies. "
        "Consider progression, timeline, and co-occurring features. "
        "Reference evidence from:",
        ["PubMed", "MedlinePlus"]
    ),
    "disease_expert": (
        "Match symptom clusters to candidate diseases and differentials. "
        "Prioritize prevalence, age group, region, and comorbidities. "
        "Reference evidence from:",
        ["Mayo Clinic", "Medscape"]
    ),
    "diagnosis_expert": (
        "Evaluate diagnostic likelihood using clinical decision rules and flowcharts. "
        "Consider Bayesian reasoning, sensitivity/specificity, and red flags. "
        "Reference evidence from:",
        ["UpToDate", "BMJ Best Practice"]
    ),
    "drug_expert": (
        "Propose medication and treatment pathways with dosing & contraindications. "
        "Consider age, severity, allergies, pregnancy, interactions, and comorbidities. "
        "Reference evidence from:",
        ["Drugs.com", "FDA Database"]
    ),
    "lab_expert": (
        "Interpret expected laboratory test findings and deviations. "
        "Consider ranges, diagnostic value, and follow-up testing. "
        "Reference evidence from:",
        ["LabCorp", "NIH"]
    ),
    "risk_expert": (
        "Estimate risk severity, complications, and escalation triggers. "
        "Include admission criteria, red-flag symptoms, and clinical thresholds. "
        "Reference evidence from:",
        ["NICE Guidelines", "CDC"]
    )
}
```

```python
def build_prompt(expert, user_input):
    base, sources = expert_prompts[expert]
    db_refs = ", ".join(sources)
    return f"{base} {db_refs}.\n\nPatient Input:\n{user_input}\n\nOutput clinically relevant reasoning:"
```

```python
prompt = build_prompt("disease_expert", "patient has fever and sore throat for 3 days")
```

```python
def get_multillm_opinion(user_input):
    outputs = {}

    for expert, model in expert_models.items():
        prompt = build_prompt(expert, user_input)
        print(f"\nðŸŸ¦ Querying {expert} ({model})...")
        out = safe_call(model, prompt)
        outputs[expert] = out
        time.sleep(1)  # avoid rate limiting

    return outputs
```

```python
def aggregate(outputs):
    text = "### AGGREGATED EXPERT REPORT\n\n"
    for expert, content in outputs.items():
        text += f"**{expert}:**\n{content}\n\n"
    return text
```

```python
def summarize_final(report):
    prompt = f"""
You are a medical consensus engine.
Given this multi-expert report, produce a final consistent medical answer with:

âœ” diagnosis likelihood
âœ” evidence
âœ” red flags
âœ” recommendations

Report:
{report}
"""

    return call_llm("deepseek/deepseek-chat", prompt)
```

```python
def safe_call(model, prompt):
    try:
        return call_llm(model, prompt)
    except Exception as e:
        print(f"âš  {model} failed, using fallback gpt-4o-mini")
        try:
            return call_llm("openai/gpt-4o-mini", prompt)
        except:
            return None
```

```python
user_case ='''High fever around 103Â°F with severe body aches, fatigue, mild cough, and chills for the past 3 days after returning from a coastal trip.
Took paracetamol twice which reduced the fever temporarily but symptoms returned. No rash or breathing difficulty,
but appetite is low and hydration is moderate. Wants to know if this could be viral fever, flu, or something related to travel and whether to visit a doctor.'''

answers = get_multillm_opinion(user_case)
report = aggregate(answers)
final = summarize_final(report)

print("\n================ FINAL CONSENSUS ================\n")
print(final)
```
